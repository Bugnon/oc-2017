The first  thing to do iy improrting the modules

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
help(cv2.imread)
```

```python
path = "images/lego briques.jpg"
img = cv2.imread(path)
```

```python
print(img.shape)
```

```python
plt.imshow(img)
```

In order to display a correct color image we have to convert BGR into RGB by using the "cv2.cvtColor(img.cv2.COLOR_BGR2RGB)"

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
def show(img):
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img2)
```

```python
show(img)
```

```python
flipped = cv2.flip(img, 0)
show(flipped)
```

```python
cropped = img[30:120 , 240:335]
show(cropped)
```
