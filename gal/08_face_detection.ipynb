```python
from __future__ import print_function
import cv2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if len(img.shape) == 3:
        img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plot = plt.imshow(img2)
    else:
        plot = plt.imshow(img, cmap = 'gray')
```

```python
img = cv2.imread('famille.jpg')
ray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(img)
```

```python
class FaceDetector:
    def __init__(self, faceCascadePath):
        self.faceCascade = cv2.CascadeClassifier(faceCascadePath)

    def detect(self, img, scaleFactor = 1.1, minNeighbors = 5,
        minSize = (30, 30)):
        rects = self.faceCascade.detectMultiScale(img,
            scaleFactor = scaleFactor,
            minNeighbors = minNeighbors, minSize = minSize,
            flags = cv2.CASCADE_SCALE_IMAGE)

        return rects
```
