# 09 Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

Inserting magic cell

```python
%matplotlib inline
```

```python
path = "../images/lego336*336.jpg"
img = cv2.imread(path)
```

```python
def show(img):
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

## Original

```python
show(img)
```

## Simple thresholding

```python
blurred = cv2.GaussianBlur(img, (5, 5), 0)
gray = cv2.cvtColor(blurred, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
```

Simple Binary Tresholding

```python
(T, thresh) = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY)
plt.imshow(thresh,cmap="gray")
```

inverse binary thresholding

```python
(T, threshInv) = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)
plt.imshow(threshInv,cmap="gray")
```

Simple Tresholding

```python
imgT=(cv2.bitwise_and(gray, gray, mask =
threshInv))
blurredT=cv2.GaussianBlur(imgT, (7,7), 0)
plt.imshow(blurredT,cmap="gray")
```

## Adaptive thresholding

Mean Thresh

```python
threshm = cv2.adaptiveThreshold(gray, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 4)
plt.imshow(threshm,cmap="gray")
```

Gaussian Tresh

```python
threshg = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 3)
plt.imshow(threshg,cmap="gray")
```

## Otsu and Riddler-Calvard

```python
import mahotas
```

Otsu’s method

```python
T = mahotas.thresholding.otsu(gray)
print("Otsu’s threshold: {}".format(T))
```

```python
thresh = gray.copy()
thresh[thresh > T] = 255
thresh[thresh < 255] = 0
thresh = cv2.bitwise_not(thresh)
plt.imshow(thresh,cmap="gray")
```

Riddler-Calvard method

```python
T = mahotas.thresholding.rc(blurred)
print("Riddler-Calvard: {}".format(T))
```

```python
thresh = gray.copy()
thresh[thresh > T] = 255
thresh[thresh < 255] = 0
thresh = cv2.bitwise_not(thresh)
plt.imshow(thresh,cmap="gray")
```
