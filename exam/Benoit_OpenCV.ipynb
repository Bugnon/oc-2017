```python
# Ensuite, il faut importer les modules nécessaires au bon fonctionnement d'OpenCV:  

%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pour faire apparraître les images on définit la fonction suivante:

def show(img):
    #vérification de la dimension de l'image (dim3=couleur; dim2=noir et blanc)
    if len(img.shape)==3: 
        img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convertion du mode de couleur
        plot = plt.imshow(img2)
    else:
        plot = plt.imshow(img, cmap = 'gray')
```

```python
# Pour faire apparaître les histograms on définit la fonction suivante:

def histogram(hist, title):
    plt.figure()
    plt.title(title)
    plt.xlabel("Bins") #Nom échelle des x
    plt.ylabel("# of Pixels") #Nom échelle des y
    plt.plot(hist)
    plt.xlim([0, 256])
    plt.show()
```

## Drawing

### Dessiner une ligne

```python
# Lire l'image avec:
img1 = cv2.imread("trex.png")
```

Ecrire la fonction pour faire apparaître une ligne:  

cv2.line(img, ($x_1,y_1$), ($x_2,y_2$), (B,G,R), e)  

img = nom de l'image sur laquelle la ligne doit apparaître  
($x_1,y_1$) = position du point (pixel) au début de la ligne

## Image processing

### Translation

Définition: une translation consiste à déplacer l'image de $t_x$ et $t_y$ pixels.

Pour faire une translation, il faut créer une matrice de translation:  

$\begin{pmatrix} 1 & 0 & t_x \\0 & 1 & t_y \end{pmatrix}$   =>    M = np.float32([ [1,0,tx] , [0,1,ty] ])  

$t_x$ = nouvelle valeur de x  
$t_y$ = nouvelle valeur de y

Ensuite il faut appeller la fonction qui réalise la translation à l'aide de la matrice:  

cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

M = nom matrice de translation  
img.shape[1] et img.shape[0] dimensions de l'image

```python
# Exemple:
img5 = cv2.imread('trex.png')

M = np.float32([ [1,0,40] , [0,1,60] ])
img5 = cv2.warpAffine(img5, M, (img.shape[1], img.shape[0]))
show(img5)

# L'image est déplacée de 45 pixels vers la droite et de 60 pixels vers le bas
```

### Rotation

Definition: la rotation consiste à tourner l'imagee de θ degrés autour d'un point.

Il faut donc définir un point de rotation:  
point = (x,y)  

Ensuite on définit la matrice de rotation:  
M = cv2.getRotationMatrix2D(point, θ, dim)  

θ: s'exprime en dergé  
dim: sont les dimensions de l'image, ainsi si l'on remplace par la valeur 1.0 l'image garde la même taille mais 2.0, elle      sera deux fois plus grande  

La fonction de rotation est la même que celle de la translation. La différence se trouve donc dans la matrice.

### Resizing 

Définition: Le resizing consiste à redéfinir les dimensions de l'image (augmenter la hauteur ou la largeur de l'image.

Il faut définir le ratio de changement:  
ex: r = n/img.shape[1] (n est un float)

Ensuite on définit les nouvelles dimensions:    
ex: dim = (n, int(img.shape[0]*r))

Enfin on appelle la fonction:  

cv2.resize(img, dim, interpolation = cv2.INTER_AREA)  

interpolation = cv2.INTER_AREA : est un algorithme de redimension  

Dans cet exemple, l'image est redimensionner en fonction de la largeur, si l'on voulait la redimensionner en fonction de la hauteur il suffirait de permuter img.shape[1] et img.shape[0] de r et dim, sans modifier la fonction.

```python
# Exemple:
img7 = cv2.imread('trex.png')

r = 50.0/img.shape[1]
dim = (50, int(img.shape[0]*r))
img7 = cv2.resize(img7, dim, interpolation = cv2.INTER_AREA)
show(img7)

#L'image a été redimensionner comme on peut voir les echelles des x et des y, elle est flou car le resizing est trop grand
```
