# Thresholding

## 9.1 simple thresholding

We must to start with importing packages

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def show(img):
    # check if the image is color (dim=3) or black and white (dim=2)
    if len(img.shape)==3:
        img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plot = plt.imshow(img2)
    else:
        plot = plt.imshow(img, cmap='gray')
```

Transforming the image

```python
path = "images/2coin.jpg"
img = cv2.imread(path)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(img2,(5,5),0)
show(img2)
```

```python
(T,thresh) = cv2.threshold(blurred,155,255,cv2.THRESH_BINARY)
show(thresh)
```

```python
(T,threshInv) = cv2.threshold(blurred,200,255,cv2.THRESH_BINARY_INV)
show(threshInv)
```

```python
coins = cv2.bitwise_and(img2,img2,mask = threshInv)
show(coins)
```
