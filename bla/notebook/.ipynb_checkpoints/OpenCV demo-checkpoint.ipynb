# OpenCV demo

The first thing to do is importing modules.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Insert a cell magic to insert images as inline elements.

```python
%matplotlib inline
```

```python
path ="../images/car.jpg"
img= cv2.imread(path)
```

```python
print(img.shape)
```

```python
plt.imshow(img)
```

In order to display a corect color image we have to convert BGR into RGB using the `cv2.cvtColor(img,cv2.Color_BGR2RGB)`

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```
