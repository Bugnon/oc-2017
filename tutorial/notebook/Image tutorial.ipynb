A short introduction on plotting images with Matplotlib.

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

## Importing image data into numpy arrays

```python
img = mpimg.imread('images/stinkbug.png')
print(img.shape)
print(img[0, 0])
```

Matplotlib has rescaled 8 bit data into float32 in the interval [0, 1]

## Plotting numpy arrays as image

```python
imgplot = plt.imshow(img)
```

```python
lum_img = img[:, :, 0]
imgplot = plt.imshow(lum_img)
```

```python
print(lum_img.shape)
```

```python
imgplot = plt.imshow(lum_img, cmap='hot')
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

## Color scale reference

```python
imgplot = plt.imshow(lum_img)
plt.colorbar()
```

## Examining a specific range 

```python
h = plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
```

```python
imgplot = plt.imshow(lum_img, clim=(0.0, 0.7))
plt.colorbar()
```

```python
fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(lum_img)
a.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
a = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(lum_img, clim=(0.0, 0.7))
a.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
plt.show()
```

## Array interpolation scheme

```python
from PIL import Image

img = Image.open('images/stinkbug.png')
img.thumbnail((64,64), Image.ANTIALIAS) # resizes image in place
imgplot = plt.imshow(img)
```

```python
imgplot = plt.imshow(img, interpolation='bicubic')
```
