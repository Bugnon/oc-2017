## Drawing Functions
The drawing methods have the following arguments
- img: the image where to draw
- color: as BGR tuple (255, 0, 0) for blue
- thickness: (-1 for filling the shape)

```python
blue = (255, 0, 0)
green = (0, 255, 0)
red = (0, 0, 255)
yellow = (0, 255, 255)
cyan = (255, 255, 0)
white = (255, 255, 255)
black = (0, 0, 0)
```

### Drawing a line

```python
img = np.zeros((256, 256, 3), np.uint8)
p0 = (0, 0)
p1 = (255, 255)
cv2.line(img, p0, p1, blue, 5)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot = plt.imshow(img2)
```

### Drawing a rectangle

```python
p0 = (20, 20)
p1 = (120, 120)
cv2.rectangle(img, p0, p1, red, 3)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot = plt.imshow(img2)
```

### Drawing a circle

```python
p0 = (120, 120)
r = 100
cv2.circle(img, p0, r, green, 2)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot = plt.imshow(img2)
```

### Drawing an ellipse

```python
center = (100, 200)
axes = (100,50)
cv2.ellipse(img, center, axes, 0, 0, 270, yellow, -1)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot = plt.imshow(img2)
```

### Drawing a polygone

```python
pts = np.array([[150, 50], [250, 50], [250, 100]], np.int32)
pts = pts.reshape((-1, 1, 2))
closed = True
cv2.polylines(img, [pts], closed, cyan, 3)
 
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot = plt.imshow(img2)
```

### Adding text to images

```python
text = 'OpenCV'
pos = (120, 180)
font = cv2.FONT_HERSHEY_SIMPLEX
size = 1
cv2.putText(img, text, pos, font, size, white, 2, cv2.LINE_AA)
 
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot = plt.imshow(img2)
```

### adding arrowed line

```python
p0 = (150, 100)
p1 = (220, 150)
cv2.arrowedLine(img, p0, p1, red, 2)
 
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot = plt.imshow(img2)
```

## Core operations

This functions allows to print an OpenCV image to the notebook.

```python
def show(img):
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plot = plt.imshow(img2)
```
