# OpenCV demo

The first thing to do is importing the modules.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Insert a cell magic to insert images as inline elements.

```python
%matplotlib inline
```

To check if the OpenCV module is present we can call the help() function.

```python
help(cv2.imread)
```

```python
path = "images/lego.jpg"
img = cv2.imread(path)
```

```python
print(img.shape)
```

```python
plt.imshow(img)
```

In order to display a correct color image we have to convert BGR into RGB by using the `cv2.cvtColor(img, cv2.COLOR_BGR2RGB)`

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
def show(img):
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img2)
```

```python
show(img)
```
