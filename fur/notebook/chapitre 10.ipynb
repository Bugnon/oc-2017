# Gradients and edge detection

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if len(img.shape)==3:
        img2 =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img2)
    else:
        plt.imshow(img, cmap="gray")
        
        
```

```python
path = "../../images/trex.png"
img= cv2.imread(path)
show(img)
```

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(gray)
```
